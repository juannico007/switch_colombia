```python
import pandas as pd
```

```python
p_t = pd.read_excel("Plantas termicas.xlsx", sheet_name = "ctermico")
g_bc_21 = pd.read_csv("gen_build_costs.csv")
```

```python
p_t.head()
```

```python
#Creación de dataframe vacio para gen_build_cost
gen_bc_t = pd.DataFrame()
```

```python
#Mapeo de generation project
gen_bc_t["GENERATION_PROJECT"] = p_t["...Nombre..."]
```

```python
#Build year - Aun no esta
num_rows = len(gen_bc_t.index)
num_rows_21 = len(g_bc_21.index)
year = []
for i in range(num_rows):
    project = gen_bc_t["GENERATION_PROJECT"][i]
    found = False
    for j in range(num_rows_21):
        if project == g_bc_21["GENERATION_PROJECT"][j]:
            year.append(g_bc_21["build_year"][j])
            found = True
            break
    if not found:
        year.append(".")
gen_bc_t["build_year"] = year
```

```python
#overnight cost - Aun no esta
oc = ["." for i in range(num_rows)]
gen_bc_t["gen_overnight_costs"] = oc
```

```python
gen_bc_t.to_csv('gen_build_costs_thermal.csv')
```
