```python
import pandas as pd
```

```python
p_te = pd.read_excel("Plantas termicas.xlsx", sheet_name = "ctermico")
g_bp_21 = pd.read_csv("gen_build_predetermined.csv")
```

```python
p_te.head()
```

```python
g_bp_21.head()
```

```python
#Seleccion de los proyectos existentes
p_t = p_te.loc[p_te["Tipo"] == 0]
#Ajuste de los indices
num_rows = len(p_t.index)
p_t = p_t.set_index([pd.Index([i for i in range(num_rows)])])
```

```python
#Creación de dataframe vacio para gen_build_predetermined
gen_bp_t = pd.DataFrame()
```

```python
#Mapeo de generation project
gen_bp_t["GENERATION_PROJECT"] = p_t["...Nombre..."]
```

```python
#Build year - Aun no esta
num_rows = len(gen_bp_t.index)
num_rows_21 = len(g_bp_21.index)
year = []
for i in range(num_rows):
    project = gen_bp_t["GENERATION_PROJECT"][i]
    found = False
    for j in range(num_rows_21):
        if project == g_bp_21["GENERATION_PROJECT"][j]:
            year.append(g_bp_21["build_year"][j])
            found = True
            break
    if not found:
        year.append(".")
gen_bp_t["build_year"] = year
```

```python
#Build gen predetermined - Aun no esta
bp = []
for i in range(num_rows):
    project = gen_bp_t["GENERATION_PROJECT"][i]
    found = False
    for j in range(num_rows_21):
        if project == g_bp_21["GENERATION_PROJECT"][j]:
            bp.append(g_bp_21["gen_predetermined_cap"][j])
            found = True
            break
    if not found:
        bp.append(".")
gen_bp_t["build_gen_predetermined"] = bp
```

```python
gen_bp_t.to_csv('gen_build_predetermined_thermal.csv')
```
