```python
import pandas as pd
```

```python
p_r1 = pd.read_excel("Plantas renovables.xlsx", sheet_name = "Plantas renovables")
p_r2 = pd.read_excel("Plantas renovables 2.xlsx", sheet_name = "cgndco")
```

```python
p_r2
```

```python
#Creación de dataframe vacio para gen_info
gen_info_r = pd.DataFrame()
```

```python
#Mapeo de generation project
gen_info_r["GENERATION_PROJECT"] = p_r2["Name"]
print(gen_info_r)
#Obtengo los proyectos que están en otro archivo
num_rows1 = len(p_r1.index)
num_rows = len(gen_info_r.index)
for i in range(num_rows1):
    found = False
    for j in range(num_rows):
        if p_r1['Nombre'][i] == gen_info_r['GENERATION_PROJECT'][j]:
            found = True
            break
    if not found:
        gen_info_r.loc[num_rows] = [p_r1['Nombre'][i]]
        num_rows += 1
```

```python
#Mapeo de tecnologías y fuente
tech = []
energy_source = []
num_rows = len(gen_info_r.index)
for i in range(num_rows):
    project = gen_info_r['GENERATION_PROJECT'][i]
    if project[0] == 'S':
        tech.append("pv_solar")
        energy_source.append('Solar')
    elif project[0] == 'E':
        tech.append('Eolica')
        energy_source.append('Wind')
    elif project[0] == 'H':
        tech.append('Hidro')
        energy_source.append('Water')
    elif project[0] == 'T':
        tech.append('Thermal')
        energy_source.append('Coal')
gen_info_r['gen_tech'] = tech
gen_info_r['gen_energy_source'] = energy_source
```
